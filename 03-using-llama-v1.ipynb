```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.llms import HuggingFaceTextGenInference

inference_server_url_local = "<ENDPOINT_IP>"

llm_local = HuggingFaceTextGenInference(
    inference_server_url=inference_server_url_local,
    max_new_tokens=400,
    top_k=10,
    top_p=0.95,
    typical_p=0.95,
    temperature=0.7,
    repetition_penalty=1.03,
)
```

```python
from langchain import PromptTemplate, LLMChain

template = """Question: {question}
Answer: Let's think step by step."""

prompt = PromptTemplate(
    template=template,
    input_variables= ["question"]
)

llm_chain_local = LLMChain(prompt=prompt, llm=llm_local)
```

```python
llm_chain_local("Can you explain briefly to me what is the Python programming language?")
```
