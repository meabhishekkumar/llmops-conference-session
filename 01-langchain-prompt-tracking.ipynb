## Prompt Tracking with LangSmith

```python
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_ENDPOINT"]="https://api.smith.langchain.com"
#os.environ["LANGCHAIN_API_KEY"]= SET THIS VALUE
os.environ["LANGCHAIN_PROJECT"]="SET_YOUR_LANGCHAIN_PROJECT"
```

```python
from dotenv import load_dotenv

from langchain.agents import AgentType, initialize_agent, load_tools
from langchain.llms import OpenAI


llm = OpenAI(model_name='text-davinci-003', temperature=0)
tools = load_tools(['wikipedia', 'llm-math'], llm=llm)
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION)
agent.agent.llm_chain.verbose=True
output = agent.run('When did India first satellite, Aryabhata launched? What is 2023 minus this year?')

print(output)
```
