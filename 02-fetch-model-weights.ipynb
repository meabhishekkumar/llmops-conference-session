```python
import gcsfs
def upload_to_gcs(src_dir: str, gcs_dst: str):
    fs = gcsfs.GCSFileSystem(project='<GCP_PROJECT_ID>', token='drive/MyDrive/gcp-service-account.json')
    fs.put(src_dir, gcs_dst, recursive=True)
```

```python
source = os.path.join(os.path.curdir, "Llama-2-7b-chat-hf")
destination = "gs://<BUCKET>/<PATH>"

upload_to_gcs(source, destination)
```
